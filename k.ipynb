```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def plot_random_images(data_dir, num_images=3):
    """
    Plot a specified number of random images from a directory.
    
    Parameters:
    data_dir (str): Path to directory containing images
    num_images (int): Number of random images to plot
    """
    # Get list of image files from directory
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff']
    image_files = []
    
    for file in os.listdir(data_dir):
        ext = os.path.splitext(file)[1].lower()
        if ext in image_extensions:
            image_files.append(os.path.join(data_dir, file))
    
    if len(image_files) < num_images:
        print(f"Warning: Only found {len(image_files)} images in the directory.")
        num_images = len(image_files)
    
    if len(image_files) == 0:
        print(f"No images found in {data_dir}")
        return
    
    # Randomly select images
    selected_images = random.sample(image_files, num_images)
    
    # Create subplot grid
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    if num_images == 1:
        axes = [axes]  # Handle case when only one image is requested
    
    # Plot each image
    for i, img_path in enumerate(selected_images):
        img = Image.open(img_path)
        axes[i].imshow(np.array(img))
        axes[i].set_title(os.path.basename(img_path))
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return selected_images

# Usage example:
# Replace 'path/to/your/data/directory' with your actual path
data_directory = 'Cattle muzzle dataset.v1i.yolov11/train/images'
selected_files = plot_random_images(data_directory)

print("Selected image files:")
for file in selected_files:
    print(f" - {file}")
```
